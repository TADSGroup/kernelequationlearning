```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.lines import Line2D

# Sample data
x = np.linspace(0, 10, 100)
y = np.sin(x)
y_lower = y - 0.2
y_upper = y + 0.2

# Define colors
line_color = "blue"
fill_color = "lightblue"

# Create plot with line and filled area
fig, ax = plt.subplots()
ax.fill_between(x, y_lower, y_upper, color=fill_color, alpha=0.5)
ax.plot(x, y, color=line_color)

# Create two legend elements for a single entry
legend_marker = Line2D(
    [0], [0], color='none', marker='s', markersize=10, 
    markerfacecolor=fill_color, alpha=0.5, markeredgewidth=0  # No border
)
legend_line = Line2D(
    [0], [0], color=line_color, linewidth=2
)

# Add custom legend with a single entry combining both elements
ax.legend(handles=[(legend_marker, legend_line)], labels=["Mean Curve with Uncertainty"], loc='upper right')

# Show plot
plt.show()
```
